```python
'''Smayan Gupta
19/17067
Section-A
Bsc Physics Hons'''
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.sparse.linalg import eigsh
```

```python
#Constants
hpi=197.3
m= 940#(0.511*10**6)
n=1000

#limits
a=5
ri=-a
rf=a
x=np.linspace(ri,rf,n)
d=((rf-ri)/n)
# anharmonic term
b=10
#harmonic term
k1=100
```

```python
#making kinetic energy operator
k=(hpi*hpi)/(2*m*d*d)

X=np.ones(n-1)
Y=np.ones(n)*(-2)  
K=np.diag(X,-1)+np.diag(Y,0)+np.diag(X,1)
K*=k
```

```python
#making harmonic potential
def Hpot(r):
    return (1/2)*k1*(r)**2

#anharmonic potential
def AHpot(r):
    return (1/2)*k1*(r)**2  + (1/3)*b*(r)**3

V=Hpot(x)
V=np.diag(V,0)

#print(V)
```

```python
H=(-K+V)
#print(H)

#energyvalues,vectors=sp.linalg.eig(H)
energyvalues,vectors=np.linalg.eigh(H)
#energyvalues,vectors=eigsh(H,k=10)
```

```python
for i in range(10,11):
    plt.plot(x,(vectors.T[i]**2),label='{} Energy level'.format(i+1))
plt.title('Wave Function for Harmonic Oscillator')
plt.grid(True)
plt.xlabel('Distance')    
plt.ylabel('Wave Function')
plt.legend()
plt.show()
```

```python
print('First Three Energy EigenValues',energyvalues[0:3])
```

```python
fig,axes= plt.subplots(1,2,figsize=(25,12))
Potentials=[Hpot(x),AHpot(x)]
title=['Harmonic Potential','Anharmonic Potential']
for i in range(2):
  ax=axes[i]
  ax.plot(x,Potentials[i])
  ax.grid(True)
  ax.set_title(title[i])
  ax.set_xlabel('Distance')    
  ax.set_ylabel('Potential')
```
